```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("Gotem_Pumpkins.csv")
test = pd.read_csv("Freyja_Pumpkins.csv")
train.head()
```

```python
train.drop(train.columns[0],axis=1,inplace=True)
test.drop(test.columns[0],axis=1,inplace=True)
test.head()
```

```python
train.dropna(inplace=True)
test.dropna(inplace=True)
```

```python
class_map={label:i for i, label in enumerate(train['Class'].unique())}
#dictionary comprehension to make sure we have labels match in training and testing data
train['Class']=train['Class'].map(class_map)
test['Class']=test['Class'].map(class_map)
```

```python
X_train=train.drop(['Class'],axis=1).values
Y_train=train['Class'].values
X_test=test.drop(['Class'],axis=1).values
Y_test=test['Class'].values
```

```python
X_train=(X_train-X_train.mean(axis=0))/X_train.std(axis=0)
X_test=(X_test-X_test.mean(axis=0))/X_test.std(axis=0)
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def loss(y,y_pred):
    return -np.mean(y*np.log(y_pred)+(1-y)*np.log(1-y_pred))

def gradient_descent(X, Y, w, b, learning_rate, iterations):
    m=X.shape[0]
    for i in range(iterations):
        fx = np.dot(X,w)+b
        y_pred=sigmoid(fx)
        dw=np.dot(X.T,(y_pred-Y))/m
        db=np.sum(y_pred-Y)/m
        w-=learning_rate*dw
        b-=learning_rate*db
        if i%100==0:
            Loss=loss(Y,y_pred)
            print(f"Iteration {i} Loss = {Loss:.4f}")
    return w,b

def predict(X, w, b):
    fx = np.dot(X, w) + b
    y_pred = sigmoid(fx)
    return np.where(y_pred > 0.5, 1, 0)
```

```python
w=np.zeros(X_train.shape[1])
b=0
learning_rate=0.01
iterations=1000
w, b = gradient_descent(X_train, Y_train, w, b, learning_rate, iterations)

y_test_pred = predict(X_test, w, b)
accuracy = np.mean(y_test_pred == Y_test) * 100

print(f"Test Accuracy: {accuracy:.2f}%")
```
